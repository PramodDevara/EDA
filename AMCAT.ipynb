```python
def Prst(x):
  if x == "present":
    return 10/6/2024
  else:
    return x
```

```python
# Converting date columns into datetime
print("Type of DOJ column = ",df["DOJ"].dtype)
df["DOJ"] = pd.to_datetime(df["DOJ"])
print("Type of DOJ column = ",df["DOJ"].dtype)
print()
print("Type of DOL column = ",df["DOL"].dtype)
df["DOL"] = df["DOL"].apply(lambda x: Prst(x))
df["DOL"] = pd.to_datetime(df["DOL"])
print("Type of DOJ column = ",df["DOJ"].dtype)
print()
print("Type of DOB column = ",df["DOB"].dtype)
df["DOB"] = pd.to_datetime(df["DOB"])
print("Type of DOJ column = ",df["DOJ"].dtype)
```

```python
df_num = df.select_dtypes(["int","float"])
df_cat = df.select_dtypes("object")
df_date = df.select_dtypes("datetime64[ns]")
print("No.of numerical columns = ",df_num.shape[1])
print("No.of object columns = ",df_cat.shape[1])
print("No.of date columns = ",df_date.shape[1])
```

```python
# Set the number of columns for the subplots
n_cols = 3
n_rows = (len(df_num.columns) + n_cols - 1) // n_cols # Calculate the number of rows needed
# Create a figure with subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, 5 * n_rows))
axes = axes.flatten() # Flatten the axes array for easy indexing
# Iterate through the columns in the DataFrame
for idx, col in enumerate(df_num.columns):
  # Use seaborn's histplot to plot histogram and KDE on the corresponding subplot
  sns.histplot(df_num[col], kde=True, color="lightblue", edgecolor='black',alpha=0.7, bins=10, ax=axes[idx])
# Set labels and title for each subplot
axes[idx].set_xlabel(col)
axes[idx].set_ylabel('Count') # Y-axis for counts
axes[idx].set_title(f'Histogram and Density Curve of {col}')
# Add mean and median lines
axes[idx].axvline(df_num[col].mean(), color='red', linestyle='dashed',linewidth=1, label='Mean')
axes[idx].axvline(df_num[col].median(), color='orange', linestyle='dashed',linewidth=1, label='Median')
axes[idx].legend()
# Remove empty subplots if any
for j in range(idx + 1, len(axes)):
  fig.delaxes(axes[j])
plt.tight_layout() # Adjust layout to prevent overlap
plt.show() # Show all plots
```

```python
#Analyzing outliers with boxplot
for i, col in enumerate(df_num.columns, 1):
  plt.figure(figsize=(8, 6))
  sns.boxplot(data=df_num, x=col, color="lightblue")
  plt.title(f'Box Plot of {col}', fontsize=16, fontweight="bold") # Apply font size and weight here
  plt.ylabel(col, fontsize=14) # Optional: adjust font size of ylabel
  plt.tight_layout()
  plt.show()
```

```python
# Define the number of rows and columns for subplots
num_columns = len(df_cat.columns)
cols = 3 # 3 columns for subplots
rows = (num_columns // cols) + (num_columns % cols > 0) # Calculate the number of rows needed
# Create a figure with subplots (3 columns and calculated rows)
fig, axes = plt.subplots(rows, cols, figsize=(15, rows * 5)) # Adjust the figsize as necessary
# Flatten the axes array for easy iteration
axes = axes.flatten()
for i, column in enumerate(df_cat.columns):
  A = df_cat[column].value_counts()
  total_count = A.sum() # Calculate the total count for percentage calculation
  # Dynamically generate the correct number of colors for the palette
  unique_categories = len(A.index[0:10]) # Get the number of unique categories for each column
  palette = sns.color_palette("Paired", unique_categories) # Generate thecorrect number of colors
  # Create the bar plot in the corresponding subplot
  bar_plot = sns.barplot(x=A.index[0:10], y=A.values[0:10], ax=axes[i], hue=A.index[0:10], palette=palette, legend=False)
  # Set the y-axis limit to increase the height of the plot
  axes[i].set_ylim(0, max(A.values[0:10]) * 1.5)
# Adding counts and percentages on top of the bars
for p in bar_plot.patches:
  count = int(p.get_height())
  percentage = (count / total_count) * 100 # Calculate percentage
  bar_plot.annotate(f'{count}',(p.get_x() + p.get_width() / 2., p.get_height()),ha='center', va='bottom', fontsize=12,fontweight="bold", rotation=90) # Rotate the label
  # Set the title ad labels for each subplot
  axes[i].set_title(f'Bar Plot for {column}', fontsize=16, fontweight="bold")
  axes[i].set_xlabel(column, fontsize=14, fontweight="bold")
  axes[i].set_ylabel('Values', fontsize=14, fontweight="bold")
  # Set x-ticks with appropriate labels, rotating them by 90 degrees
  axes[i].set_xticks(range(len(A.index[0:10]))) # Set ticks
  axes[i].set_xticklabels(A.index[0:10], rotation=90) # Rotate x-tick labels to 90 degrees
# Hide any unused subplots if the number of columns is not a perfect multiple of rows
for j in range(i + 1, len(axes)):
  fig.delaxes(axes[j])
plt.tight_layout()
plt.show()
```

```python
from scipy.stats import chi2_contingency
contingency_table = pd.crosstab(df['Gender'], df['Branch'], margins=True)
chi2, p, dof, expected = chi2_contingency(contingency_table.iloc[:-1, :-1])
plt.figure(figsize=(12, 6))
sns.countplot(data=df, y='Branch', hue='Gender', order=df['Branch'].value_counts().index)
plt.title('Distribution of Specialization by Gender', fontsize=16)
plt.xlabel('Specialization', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend(title='Gender')
plt.show()
```
